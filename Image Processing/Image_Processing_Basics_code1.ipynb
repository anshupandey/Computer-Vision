# Color Detection

```python
def detect_red(img):
    rows,cols,colors = img.shape
    img2 = img.copy()
    for i in range(rows):
        for j in range(cols):
            if img[i,j,0]>150 and img[i,j,1]<120 and img[i,j,2]<120:
                img2[i,j,0] = 255
                img2[i,j,1] = 255
                img2[i,j,2] = 255
    return img2
```

# Video Processing

```python
#this code will not work in google colab
vid = cv2.VideoCapture(0)
while True:
    flag,img = vid.read()
    img2 = detect_red(img)
    cv2.imshow("Video",img2)
    #to close video
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
vid.release()
cv2.destroyAllWindows()
```
